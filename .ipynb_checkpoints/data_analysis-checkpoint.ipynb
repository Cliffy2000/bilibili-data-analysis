```python
import jieba
import datetime
import openpyxl
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
from snownlp import SnowNLP as nlp
```

```python
outer = pd.read_excel('外页数据.xlsx')
inner1 = pd.read_excel('内页数据1.xlsx')
inner2 = pd.read_excel('内页数据2.xlsx')
comments_in = pd.read_excel('弹幕信息.xlsx')
comments_after = pd.read_csv('bilibili评论的数据.csv', encoding='utf-8', engine='python')
```

```python
file_names = ['外页数据.xlsx', '内页数据1.xlsx', '内页数据2.xlsx', '弹幕信息.xlsx', 'bilibili评论的数据.csv']
file_reads = [outer, inner1, inner2, comments_in, comments_after]

for i in range(5):
    print('========', str(file_names[i]), '========\n', sep='')
    file_reads[i].info()
    print('\n\n\n')
```

```python
count = 0
title_emotion_sum = 0
title_emotion_list = []
for title in outer['标题']:
    try:
        nlpObj = nlp(title)
        title_emotion = nlpObj.sentiments
        count += 1
        title_emotion_sum += title_emotion
        title_emotion_list.append(title_emotion)
        if count%5 == 0:
            print('Loading', '·'*int((count/5)%5), sep='', end='\r')
        if count%25 == 0:
            print('Loading        ', end='\r')
    except:
        continue
print('Title emotion calculations complete. ')
```

```python
def check_emotion(s):
    nlpObj = nlp(s)
    score = nlpObj.sentiments
    return score
```

```python
for i in range(25):
    title = outer['标题'][i]
    print(check_emotion(title))
```

```python
title_emotion_sum / count
```

```python
print(check_emotion('我很快乐'))
print(check_emotion('我很快乐!!'))
```

```python
title_trimmed = []
count = 0
punctuations = '，。！@#￥%……&*（）——+-=【】、{}|；：’‘“”《》？/'
for title in outer['标题']:
    new = ''
    for char in title:
        if char in punctuations:
            continue
        new += str(char)
        
    title_trimmed.append(new)
    count += 1
    print('Current progress:', count, end='\r')
```

```python
count = 0
title_emotion_sum_trimmed = 0
title_emotion_list_trimmed = []
for title in title_trimmed:
    try:
        nlpObj = nlp(title)
        title_emotion_trimmed = nlpObj.sentiments
        count += 1
        title_emotion_sum_trimmed += title_emotion_trimmed
        title_emotion_list_trimmed.append(title_emotion_trimmed)
        if count%5 == 0:
            print('Loading', '·'*int((count/5)%5), sep='', end='\r')
        if count%25 == 0:
            print('Loading        ', end='\r')
    except:
        continue
print('Title emotion calculations complete. ')
```

```python
title_emotion_sum_trimmed / count
```
